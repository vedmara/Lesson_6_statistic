```python
#Найдём какая модель самая популярная
sns.countplot(y = 'model', data = audi, order = audi['model'].value_counts().nlargest(15).index)
plt.title("The most popular model")
plt.ylabel('model')
plt.xlabel('No. of Vehicles');
plt.show()
```

```python
Из графика можно заключить, что лидирующие позиции занимают A3, Q3, A4, A1. Причем, A3 вырывается в лидеры
```

```python
#Выясним как автомобили разделяются по трансмиссии
audi['transmission'].value_counts().plot.bar()                     
plt.title("Cars sold by transmission")
plt.ylabel('Number of vehicles')
plt.xlabel('Transmission Tipe');
```

```python
#Выясним как автомобили разделяются по трансмиссии в процентном соотношении
(audi['transmission'].value_counts() / len(audi)*100).plot.bar()               
plt.title("Transmission type diagram")
plt.ylabel('% of vehicles')
plt.xlabel('Transmision type');
```

```python
#Выясним как автомобили разделяются по типу двигателя в процентном соотношении
(audi['fuelType'].value_counts() / len(audi)*100).plot.bar()               
plt.title("Fuel type diagram")
plt.ylabel('% of vehicles')
plt.xlabel('Fuel type');
```

```python
#Выясним как автомобили разделяются по объему двигателя в процентном соотношении
(audi['engineSize'].value_counts() / len(audi)*100).plot.bar()               
plt.title("Engine size diagram")
plt.ylabel('% of vehicles')
plt.xlabel('Engine size');
```

```python
# Анализ соотношения Типа коробки передач и Объёма двигателя
sns.countplot(y='engineSize',data=audi, hue='transmission')
plt.title("Engine size v/s Transmission")
plt.ylabel('Transmission')
plt.xlabel('Count of vehicles');
#Самый популярный 2х литровый двигатель представлен на рынке со всеми типами управления, но вперед вырывается тип ручного управления 
```

```python
# Создадим новую колонку 'price_group' где сгруппируем цены за автомобили
audi['price_group'] = pd.cut(audi['price'],[0,2000,5000,10000,30000,50000,150000], 
                                labels = ['<2K','2-5K','5-10K','10-30K','30-50K','>50K'], include_lowest = True)
audi['price_group'] = audi['price_group'].astype(object)
(audi['price_group'].value_counts() / len(audi) *100).plot.bar()
plt.title("Price Group bar diagram")
plt.ylabel('% of vehicles')
plt.xlabel('Price Group');
#Основной поток машин продается в пределах 10-30 тысяч фунтов
```

```python
# Модель автомобиля марки Audi к цене
sns.countplot(y='price_group',data=audi, hue='model', order=['<2K','2-5K','5-10K','10-30K','30-50K','>50K'])           
plt.title("Price v/s Model")
plt.ylabel('Price Group')
plt.xlabel('Count of vehicles');
```

```python
#Мы видим, что наиболее популярные модели автомобиля находятся в пределах 10-30 тысяч фунтов
```

```python
#Выясним как объем двигателя (engineSize) соотносится с пробегом автомобиля?

audi[audi['engineSize'] < 6].plot.hexbin(x='engineSize', y='mileage', sharex=False, gridsize=12)
plt.title("Engine Size  v/s Mileage")
plt.ylabel('Mileage')
plt.xlabel('Engine Size');
```

```python
#Исходя из этой диаграммы мы можем сказать, что большинство автомобилей продаются с объемом двигателя 2 литра и пробегом до 25000.
```

```python
# Выясним как цена относится к возрасту авто
audi[audi['price'] <= 150000].plot.scatter(x='year',y='price')
plt.title("Price v/s Year")
plt.ylabel('Price')
plt.xlabel('Year');

```

```python
car_corr = audi.corr()
car_corr
plt.figure(figsize=(5,5))
sns.heatmap(car_corr,vmax=1,square=True,annot=True, cmap='BuGn')
plt.title("Correlation between various features of car")
plt.show()
```

```python
По этой тепловой карте мы можем сказать, что:
    1) переменные цены и года выпуска имеют сильную корреляцию, то есть более новые автомобили имеют высокую цену, а автомобили старые имеют более низкую цену.
    2) цена и объем двигателя имеют слабую положительную корреляцию.
    3) объем двигателя и пробег также имеют некоторую слабую положительную корреляцию, объем двигателя и пробег не имеют положительной корреляции, 
    но по моим данным большинство автомобилей имеют объем двигателя 2 литра, из-за чего пробег и объем двигателя имеют положительную корреляцию.
    4) переменные года выпуска и пробега имеют сильную корреляцию, то есть автомобили старые имеют большой пробег, а автомобили более новые имеют низкий пробег.
    5) цена и пробег имеют отрицательную корреляцию, то есть автомобили с более низкой ценой имеют более высокий пробег, а автомобили с более высокой ценой имеют более низкий пробег.
    
```
